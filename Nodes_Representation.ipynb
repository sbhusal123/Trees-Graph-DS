# Representation of a Node In Binary Tree
- Left SubTree
- Right SubTree
- Data Value

# Constraints 
- Left child always lesser than parent.
- Right Child is always greater than parent

node.jpeg.drawio.png

```python
class Node:

  def __init__(self, data, left=None, right=None):
    self.left = left
    self.right = right
    self.data = data
```

# Representing Above Tree

```python
# Representation of Above tree

tree = Node(12, # root of BT

            # parent is 12 
            Node(10, # 10 < 12 (left to 12)
              
                    # parent is 10
                    Node(9), # 9 < 10 (left to 10)

                    # parent is 10  
                    Node(11) # 11 > 10 (right to 10)
                 ),

            # parent is 12
            Node(15, # 15 > 12 (right to 12)

                    # parent is 15
                    Node(14), # 14 < 15 (left to 15)

                    # parent is 15
                    Node(16) # 16 > 15 (right to 15)
                 )
          )
```
